## Log Analysis

Analyze structured logs from the example scenarios.

```python
# Query logs for our test scenarios
print("\n" + "="*70)
print("LOG ANALYSIS - Recent Events")
print("="*70)

# Get all logs from last hour
recent_logs = logger.query_logs(
    start_time=datetime.now() - timedelta(hours=1),
    limit=50
)

print(f"\nðŸ“Š Total Events: {len(recent_logs)}")

# Group by event type
event_counts = {}
for log in recent_logs:
    event_type = log.get('event_type', 'unknown')
    event_counts[event_type] = event_counts.get(event_type, 0) + 1

print("\nðŸ“ˆ Events by Type:")
for event_type, count in event_counts.items():
    print(f"  {event_type}: {count}")

# Show routing decisions
print("\nðŸ”„ Routing Decisions:")
routing_logs = [log for log in recent_logs if log.get('event_type') == 'routing']
for log in routing_logs[:5]:  # Show first 5
    details = log.get('details', {})
    print(f"  {details.get('from_agent')} â†’ {details.get('to_agent')}: {details.get('reason')}")

# Show classifications
print("\nðŸ“‹ Classifications:")
classification_logs = [log for log in recent_logs if log.get('event_type') == 'classification']
for log in classification_logs[:3]:  # Show first 3
    details = log.get('details', {})
    classification = details.get('classification', {})
    print(f"  Thread {log.get('thread_id')}: {classification.get('issue_type')} | {classification.get('urgency')} | Confidence: {details.get('confidence', 0):.2f}")
```

```python
# Get comprehensive metrics
print("\n" + "="*70)
print("PERFORMANCE METRICS")
print("="*70)

metrics = logger.get_metrics(time_window_hours=24)

print(f"\nðŸ“Š Summary:")
print(f"  Total Tickets: {metrics['total_tickets']}")
print(f"  Total Events: {metrics['total_events']}")
print(f"  Escalations: {metrics['escalations']}")
print(f"  Errors: {metrics['errors']}")
print(f"  Tool Executions: {metrics['tool_executions']}")

print(f"\nâœ… Resolution Stats:")
resolutions = metrics['resolutions']
print(f"  Successful: {resolutions['success']}")
print(f"  Failed: {resolutions['failed']}")
print(f"  Avg Confidence: {resolutions['avg_confidence']:.2f}")

if resolutions['success'] + resolutions['failed'] > 0:
    resolution_rate = resolutions['success'] / (resolutions['success'] + resolutions['failed']) * 100
    print(f"  Resolution Rate: {resolution_rate:.1f}%")

print(f"\nðŸ“ˆ Events by Agent:")
for agent, count in metrics['events_by_agent'].items():
    print(f"  {agent}: {count}")
```

## View Conversation History

Retrieve conversation history from checkpointer for any thread.

```python
# View conversation history for a specific thread
thread_id = "TEST-001"  # Change to view different thread

try:
    history = list(orchestrator.get_state_history(
        config = {
            "configurable": {
                "thread_id": thread_id,
            }
        }
    ))
    
    if history:
        print(f"\nðŸ“œ Conversation History for Thread: {thread_id}")
        print("="*70)
        
        messages = history[0].values.get("messages", [])
        for i, msg in enumerate(messages, 1):
            role = "User" if msg.__class__.__name__ == "HumanMessage" else "Assistant"
            content = msg.content[:200] + "..." if len(msg.content) > 200 else msg.content
            print(f"\n{i}. {role}:")
            print(f"   {content}")
    else:
        print(f"No history found for thread: {thread_id}")
        
except Exception as e:
    print(f"Error retrieving history: {e}")
```
